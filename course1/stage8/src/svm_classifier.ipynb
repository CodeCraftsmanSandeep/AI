```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../stage7/data/Movie_classification.csv", header = 0)
```

```python
df.head()
```

```python
df.info()
```

# Data pre-processing

### Missing value imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### Dummy variable creation

```python
df = pd.get_dummies(df, columns = ['3D_available', 'Genre'], drop_first = True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Convert boolean columns to integer
df = df.astype({col: 'int64' for col in df.select_dtypes(include='bool').columns})
```

```python
df.info()
```

### X-y  split

```python
X = df.loc[:, df.columns != 'Start_Tech_Oscar']
```

```python
X.info()
```

```python
y = df['Start_Tech_Oscar']
```

```python
y
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.head()
```

```python
X_train.shape, X_test.shape
```

# Classification

### Standardizing

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_test_std
```

### Training SVM

```python
from sklearn import svm
```

```python
clf_svm_1 = svm.SVC(kernel = 'linear', C = 0.01) # classification
clf_svm_1.fit(X_train_std, y_train)
```

### Predicting values

```python
y_train_pred = clf_svm_1.predict(X_train_std)
y_test_pred = clf_svm_1.predict(X_test_std)
```

```python
y_test_pred
```

### Model performance

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
confusion_matrix(y_test, y_test_pred) # confusion matrix on test data set
```

```python
accuracy_score(y_test, y_test_pred) # accuracy score
```

```python
clf_svm_1.n_support_ # number of support vectors
# first number - number of support vectors for 1st class
# second number - number of support vectors for the second class
```

### Grid search

```python
# Hyper value tuning
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'C' : (0.01, 0.005, 0.02, 1, 3, 4, 5, 6, 7, 8, 9, 100, 11, 203, 1, 0.5, 0.6)
}
```

```python
clf_svm_1 = svm.SVC(kernel = 'poly')
```

```python
svm_grid_lin = GridSearchCV(clf_svm_1, params, n_jobs = -1, cv = 10, verbose = 1, scoring = 'accuracy')
```

```python
svm_grid_lin.fit(X_train_std, y_train)
```

```python
svm_grid_lin.best_params_
```

```python
linsvm_clf = svm_grid_lin.best_estimator_ # betting best model
```

```python
accuracy_score(y_test, linsvm_clf.predict(X_test_std))
```

### Polynomial kerenel

```python
clf_svm_p3 = svm.SVC(kernel = 'poly', degree = 4, C = 100)
clf_svm_p3.fit(X_train_std, y_train)
```

```python
y_train_pred = clf_svm_p3.predict(X_train_std)
y_test_pred = clf_svm_p3.predict(X_test_std)
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
clf_svm_p3.n_support_
```

### Radial kernel

```python
clf_svm_r = svm.SVC(kernel = 'rbf', gamma = 0.5, C = 10)
clf_svm_r.fit(X_train_std, y_train)
```

```python
y_train_pred = clf_svm_r.predict(X_train_std)
y_test_pred = clf_svm_r.predict(X_test_std)
```

```python
accuracy_score(y_test_pred, y_test)
```

```python
accuracy_score(y_train_pred, y_train)
```

```python
clf_svm_r.n_support_
```

### Radial Grid

```python
params = {
    'C' : (0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50),
    'gamma' : (0.001, 0.01, 0.1, 0.5, 1)
}
```

```python
clf_svm_r = svm.SVC(kernel = 'rbf')
```

```python
svm_grid_rad = GridSearchCV(clf_svm_r, params, n_jobs = -1, cv = 3, verbose = 1, scoring = 'accuracy')
```

```python
svm_grid_rad.fit(X_train_std, y_train)
```

```python
best_model = svm_grid_rad.best_estimator_
```

```python
accuracy_score(y_test, best_model.predict(X_test_std))
```
