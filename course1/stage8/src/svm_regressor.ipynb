```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importing and pre-processing data

```python
df = pd.read_csv("../../stage7/data/Movie_regression.csv", header = 0)
```

```python
df.head()
```

```python
df.info()
```

## Missing value imputation

```python
# time taken variable has missing values in it
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
# see that there are no missing values
```

## Dummy variable creation

```python
df.head()
```

```python
# there are 2 categorical variables
# 1) 3D_available
# 2) Genre
```

```python
df = pd.get_dummies(df, columns = ['3D_available', "Genre"], drop_first = True) # drop_first = True will drop the extra dummy variable
```

```python
df.head()
```

```python
df.info()
```

```python
# Convert boolean columns to integer
df = df.astype({col: 'int64' for col in df.select_dtypes(include='bool').columns})
```

```python
df.info()
```

## X-y split

```python
# target variable is collection variable
```

```python
X = df.loc[:, df.columns != "Collection"]
```

```python
type(X)
```

```python
X.shape
```

```python
y = df['Collection']
```

```python
type(y)
```

```python
y.shape
```

```python
y
```

```python
X
```

```python
df
```

## Test train split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

## Standardizing data

```python
# converting all variable such that mean is zero and standard deviation is 0 and all values are in between -1 to 1
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler().fit(X_train)
```

```python
# In svm it is must to standardize because we are calculating distance to get the model
```

```python
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
X_test_std
```

```python
X_test
```

## Training Regression Tree

```python
from sklearn.svm import SVR
svr = SVR(kernel = 'linear', C = 500)
# C is the penality parameter
```

```python
svr.fit(X_train_std, y_train)
```

## Predict values using trained model

```python
y_test_pred = svr.predict(X_test_std)
y_train_pred = svr.predict(X_train_std)
```

```python
y_test_pred
```

## Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)   
```

```python
r2_score(y_test, y_test_pred)
```
