```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/Movie_classification.csv", header = 0)
```

```python
df.head()
```

```python
df.info()
```

# Data pre-processing

### Missing value imputation

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### Dummy variable creation

```python
df = pd.get_dummies(df, columns = ['3D_available', 'Genre'], drop_first = True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Convert boolean columns to integer
df = df.astype({col: 'int64' for col in df.select_dtypes(include='bool').columns})
```

```python
df.info()
```

### X-y  split

```python
X = df.loc[:, df.columns != 'Start_Tech_Oscar']
```

```python
X.info()
```

```python
y = df['Start_Tech_Oscar']
```

```python
y
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.head()
```

```python
X_train.shape, X_test.shape
```

# Classification Tree

### Training Classification Tree

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier(max_depth = 3) # by default the criteris is gini
```

```python
clftree.fit(X_train, y_train)
```

### Predicting dependent variable value

```python
y_train_pred = clftree.predict(X_train)
y_test_pred = clftree.predict(X_test)
```

```python
y_test_pred
```

### Model performance

```python
# for regression we used r2_score and mean_squared_error
# for classification, let us use confusion_matrix and accuracy_score
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_train, y_train_pred))
print("Training error = ", 1-accuracy_score(y_train, y_train_pred))
```

```python
print(confusion_matrix(y_test, y_test_pred))
print("Test error = ", 1-accuracy_score(y_test, y_test_pred))
```

### plot decision tree

```python
dot_data = tree.export_graphviz(clftree, out_file = None, feature_names = X_train.columns, filled = True)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())

# orange - negatives dominate
# blue - positive dominate
# white - no purity
```

### Controlling tree growth of the classification tree

```python
clftree2 = tree.DecisionTreeClassifier(min_samples_leaf = 10, max_depth = 10)
clftree2.fit(X_train, y_train)

dot_data = tree.export_graphviz(clftree2, out_file = None, feature_names = X_train.columns, filled = True)
from IPython.display import Image
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
print("Test error = ", 1 - accuracy_score(y_test, clftree2.predict(X_test)))
```
